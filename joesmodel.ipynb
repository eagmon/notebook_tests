```python
from bigraph_viz import plot_bigraph
plot_settings = {'size':'8,5', 'plot_schema': True}
```

```python
Fe_schema = {
        'loc1': {
            '_value': 1.0, '_type': 'pos',
            'Fe': {'_value': 0.1, '_type': 'conc'}
        },
        'loc2': {
            '_value': 2.0,
            'Fe': {'_value': 0.2, '_type': 'conc'}
        }
    
    }
cell_schema = {
            'y': {},
            'Fe_L': {},
            'Fe_B': {},
            'internal': {},
            'COPASI': {
                '_ports': {
                    'internal': {},
                    'lumen': {},
                    'blood': {}
                },
                '_wires': {
                    'internal': 'internal',
                    'lumen': 'Fe_L',
                    'blood': 'Fe_B',
                },
            }
        }

bigraph = {
    'lumen': Fe_schema,
    'blood': Fe_schema,
    'value from location (lumen)': {
        '_ports': {
            'cell loc and conc': {'*': {}},
            'lumen loc': {},
        },
        '_wires': {
            'cell loc and conc': 'cells',
            'lumen loc': 'lumen',
        }
    },
    'cells': {
        '1': cell_schema,
        '2': cell_schema,
    },
    'migrate': {
        '_ports': {'cell locs': {}},
        '_wires': {'cell locs': 'cells'}
    }
}

# plot
plot_bigraph(bigraph, **plot_settings, out_dir='out' ,filename='joesmodel')
```
