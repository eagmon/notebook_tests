What models?
- representations of real-world systems
  - physical models: small reconstruction of a larger object,
  - biological models: like mouse, zebrafish. Animals that represent a phenomenon in humans or otherwise.
  - mathematical models: formal system with symbols and rules of operation, which represent a real world system, and whose operation corresponds to behavior of that system.
  - computational model: has algorithms that operate on a model and solve it, rather than being done analytically. Get more high-throughput due to automation.

What are simulations?
- a computational model executed through time.

 ## mathematical model

LaTeX

$
\frac{dP}{dt} = r*P
$

- $P$ represents population size
- $r$ is a rate of change, growth rate
- $t$ is time, $dt$ is the timestep



```python
def population_growth(P, r):
    dPdt =  r * P
    return dPdt 
```

```python
P0 = 110  # initial population
r = 0.1

# solve the equation
dPdt = population_growth(P0, r)

# print the result
print(f'dPdt = {dPdt}')   # this is an f-string
```

# simulate with Euler integration

```python
total_time = 10
dt = 0.1

# initial state
P = 10
r = 0.1
t = 0

# list to track history
P_history = [P]
time_history = [t]

# run the simulation
while t < total_time:

    # use function to get population change
    dPdt = population_growth(P, r)

    # grow the population
    P += dPdt * dt

    # update time
    t += dt
    t = round(t, 1)

    # save to history
    P_history.append(P)
    time_history.append(t)
```

## plot the results

```python
plt.plot(time_history, P_history)
plt.xlabel('time (years)')
plt.ylabel('population (millions of counts)')
plt.title('exponential population growth')
```
